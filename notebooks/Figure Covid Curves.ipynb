```python
from data_scientia.data import municipios
from data_scientia.features import covid_municipalities

municipio_codes_mexico = municipios.get_municipio_codes(state_name='MÃ©xico')
covid_data = covid_municipalities.get_daily_cases(
    municipio_codes=list(municipio_codes_mexico.keys())[:10])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(18, 4))

covid_data.rename(
    columns=municipio_codes_mexico
).loc[
    '2020-04-01':'2020-11-01'
].plot(
    grid=True,
    cmap='tab20',
    linewidth=2,
    alpha=.8,
    ax=ax)
ax.set_title('Daily COVID-10 cases per Municipality', fontsize=15)
ax.legend(
    loc='upper center', bbox_to_anchor=(0.5, -0.07), fancybox=True, shadow=True, ncol=5, fontsize=10)
```
